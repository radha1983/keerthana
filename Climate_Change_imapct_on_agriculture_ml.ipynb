**DATA VISUALIZATION**

1. BAR PLOT

```python
#......AVERAGE TEMPERATURE BY REGION......
#   BAR PLOT
plt.figure(figsize=(12, 6)) # Wide and Tall measurements
sns.barplot(x='Region', y='Average_Temperature_C', data=data, estimator='mean')
#estimator=mean --calculates the mean and then plot in y axis
plt.title('Average Temperature by Region')
plt.xlabel('Region')
plt.ylabel('Average Temperature (Â°C)')
plt.xticks(rotation=45) #the label is tilted to 45 degree for easy readability
plt.show()
```

<br>**LINE PLOT**

```python
#CROP YEILD OVER THE YEARS
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Crop_Yield_MT_per_HA', data=data) #Crop Yield in metric tons(no of crops)per hectare((land measurents).
plt.title('Crop Yield Over Years')
plt.xlabel('Year')
plt.ylabel('Crop Yield (MT per HA)')
plt.show()
```

Is the crop yield increasing, decreasing, or staying stable?

A rising line suggests improvement in farming techniques, technology, or favorable conditions.

A falling line could indicate climate issues, soil degradation, or lack of resources.<br><br>
ð Lowest Crop Yield was in **2024**, with an average of 2.13 MT/HA

ð Highest Crop Yield was in **2008**, with an average of 2.39 MT/HA
<br><br>
Even though we might expect an increasing trend, the data shows a decline from the peak in 2008 to a low in 2024. This could indicate:

a)Negative impacts of climate change (e.g., extreme weather, drought)<br>
b)Soil degradation( decline in soil health and quality)

<BR>**HISTOGRAM**

```python
# distribution of CO2 EMISSION

plt.figure(figsize=(12, 6))
sns.histplot(data['CO2_Emissions_MT'], bins=30, kde=True)
plt.title('Distribution of CO2 Emissions')
plt.xlabel('CO2 Emissions (MT)')
plt.ylabel('Frequency')
plt.show()
```

1. Right-Skewed Distribution (Long Tail to the Right)
    * Right-skewed means the bulk of the data is on the lower end, with a few extreme

    * It implies most regions emit low to moderate COâ, but a few emit much more
2. High Frequency on the Left (Mode of the Data)

    *  In the 10â20 metric tons (MT) range.
    *  The tallest bars in the histogram show where the most common values lie.

```python
#CROP yEILD WITHIN A COUNTRY

plt.figure(figsize=(12, 6))
sns.boxplot(x='Country', y='Crop_Yield_MT_per_HA', data=data)
plt.title('Crop Yield by Country')
plt.xlabel('Country')
plt.ylabel('Crop Yield (MT per HA)')
plt.xticks(rotation=90)
plt.show()
```

<BR> CORRELATION MATRIX

```python
plt.figure(figsize=(12, 10))
correlation_matrix = data[['Average_Temperature_C', 'Total_Precipitation_mm', 'CO2_Emissions_MT', 'Crop_Yield_MT_per_HA',
                           'Extreme_Weather_Events', 'Irrigation_Access_%', 'Pesticide_Use_KG_per_HA', 'Fertilizer_Use_KG_per_HA',
                           'Soil_Health_Index', 'Economic_Impact_Million_USD']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```

```python
sns.pairplot(data[['Average_Temperature_C', 'Total_Precipitation_mm', 'CO2_Emissions_MT', 'Crop_Yield_MT_per_HA',
                   'Extreme_Weather_Events', 'Irrigation_Access_%', 'Pesticide_Use_KG_per_HA',
                   'Fertilizer_Use_KG_per_HA', 'Soil_Health_Index']])
plt.title('Pairwise Relationships in Data')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Region', y='Economic_Impact_Million_USD', data=data)
plt.title('Economic Impact by Region')
plt.xlabel('Region')
plt.ylabel('Economic Impact (Million USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Crop_Type', y='Irrigation_Access_%', data=data)
plt.title('Irrigation Access by Crop Type')
plt.xlabel('Crop Type')
plt.ylabel('Irrigation Access (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Region', hue='Extreme_Weather_Events', data=data)
plt.title('Extreme Weather Events by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Pesticide_Use_KG_per_HA', data=data)
plt.title('Pesticide Use Over Years')
plt.xlabel('Year')
plt.ylabel('Pesticide Use (KG per HA)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['Soil_Health_Index'], bins=30, kde=True)
plt.title('Distribution of Soil Health Index')
plt.xlabel('Soil Health Index')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
temp_yield_corr = data[['Average_Temperature_C', 'Crop_Yield_MT_per_HA']].corr()
sns.heatmap(temp_yield_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation between Temperature and Crop Yield')
plt.show()
```
